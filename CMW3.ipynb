Plot all the graph from Q no 3 on the same pane using subplot.

```python
import matplotlib.pyplot as plt
import math

x = list(range(20))

# Functions from Q3
y1 = [5*(i**3) + 2*i - 1 for i in x]
y2 = [-2*(i**3) + (i**2) + 100 for i in x]
y3 = [2*math.pi*i + 20 for i in x]

# Subplot 1
plt.subplot(3, 1, 1)
plt.plot(x, y1)
plt.title("f(x) = 5x^3 + 2x - 1")

# Subplot 2
plt.subplot(3, 1, 2)
plt.plot(x, y2)
plt.title("f(x) = -2x^3 + x^2 + 100")

# Subplot 3
plt.subplot(3, 1, 3)
plt.plot(x, y3)
plt.title("f(x) = 2πx + 20")

# Adjust layout so titles don’t overlap
plt.tight_layout()

plt.show()
```

Plot COS function (-3π to 3π)

```python
import math

x = []
y = []

i = -3*math.pi
while i <= 3*math.pi:
    x.append(i)
    y.append(math.cos(i))
    i += 0.1

plt.plot(x, y)
plt.title("cos(x)")
plt.show()
```

Circle using while loop

```python
import math

x = []
y = []

a = 2
b = 2
r = 4

theta = 0
while theta <= 2*math.pi:
    x.append(r*math.cos(theta) + a)
    y.append(r*math.sin(theta) + b)
    theta += 0.1

plt.plot(x, y)
plt.gca().set_aspect('equal')
plt.show()
```

Ellipse

```python
x = []
y = []

a = 3
b = 5
center = [2,2]

theta = 0
while theta <= 2*math.pi:
    x.append(a*math.cos(theta) + center[0])
    y.append(b*math.sin(theta) + center[1])
    theta += 0.1

plt.plot(x, y)
plt.gca().set_aspect('equal')
plt.show()
```

Circle + Ellipse together

```python
import math

x1,y1 = [],[]
x2,y2 = [],[]

theta = 0
while theta <= 2*math.pi:
    x1.append(4*math.cos(theta)+2)
    y1.append(4*math.sin(theta)+2)

    x2.append(3*math.cos(theta)+2)
    y2.append(5*math.sin(theta)+2)

    theta += 0.1

plt.plot(x1,y1,label="Circle")
plt.plot(x2,y2,label="Ellipse")
plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```

Parabola

```python
import matplotlib.pyplot as plt

# Parameters that can be changed
p = 2
q = 3
a = 1

# Generate x values
x = []
y = []

i_min = -5+p
i_max = 5+p
while i_min <= i_max:
    x.append(i_min)
    y.append(4*a*(i_min- p)**2+ q)
    i_min += 0.1

# Plot
plt.plot(x, y)
plt.title(f"Parabola with vertex at ({q},{p})")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()

plt.show()
```
