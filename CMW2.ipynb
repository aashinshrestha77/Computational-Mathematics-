Write a Python function that takes a list and returns a new list with distinct elements from the
first list.

```python
def unique_list(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique
sample = [2,2,5,7,8,3,9,3,1,3,2]
print("Unique List:", unique_list(sample))
```

Write a Python function that takes a number as a parameter and checks whether the number
is prime or not.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
num = 17
if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
def twin_primes(limit=1000):
    for i in range(3, limit - 2):
        if is_prime(i) and is_prime(i + 2):
            print(i, "and", i + 2)
twin_primes()
```

A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total
n = 9
print("Sum of divisors of", n, "is:", sumOfDivisors(n))
```

Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
def mean_median(lst):
    total = 0
    for num in lst:
        total += num
    mean = total / len(lst)

    sorted_lst = sorted(lst)
    n = len(sorted_lst)

    if n % 2 == 1:
        median = sorted_lst[n // 2]
    else:
        median = (sorted_lst[n // 2 - 1] + sorted_lst[n // 2]) / 2

    return mean, median
marks = [20, 30, 35, 55, 70, 80, 90]
mean, median = mean_median(marks)
print("Marks:", marks)
print("Mean:", mean)
print("Median:", median)
```

Write a function that converts a decimal number to binary number.

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary
num = 10
print("Decimal:", num)
print("Binary:", decimal_to_binary(num))
```

Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(n):
    product = 1
    while n > 0:
        digit = n % 10
        product *= digit
        n //= 10
    return product
num = 1234
print("Number:", num)
print("Product of digits:", prodDigits(num))
```

Write a function that will take in an unknown number of arguments and multiply all of them
together and run the function for these 2 sets of numbers:

a. 1,2,3,4,5

b. 12,13,14

```python
def multiply_all(**kwargs):
    result = 1
    for value in kwargs.values():
        result *= value
    return result

# 1, 2, 3, 4, 5
print(multiply_all(a=1, b=2, c=3, d=4, e=5))

# 12, 13, 14
print(multiply_all(x=12, y=13, z=14))
```

some simple operations on set:

```python
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}

# a. Length of set1
print("a:", len(set1))

# b. Sum of elements in set2
print("b:", sum(set2))

# c. Union
print("c:", set1 | set2)

# d. Intersection
print("d:", set1 & set2)

# e. set1 - set2
print("e:", set1 - set2)

# f. set2 - set1
print("f:", set2 - set1)

# g. Check subset
print("g: set3 subset of set1:", set3.issubset(set1))
print("   set3 subset of set2:", set3.issubset(set2))

# h. Complement of set1 (U = {1 to 50})
U = set(range(1, 51))
print("h:", U - set1)

# i. Check if 15 is in set1
print("i:", 15 in set1)

# j. Add 60 to set1
set1.add(60)
print("j:", set1)

# k. Remove 60 from set2
set2.discard(60)
print("k:", set2)

# l. Remove all elements from set3
set3.clear()
print("l:", set3)

# m. Max and Min of set2
print("m: max =", max(set2), ", min =", min(set2))
```
