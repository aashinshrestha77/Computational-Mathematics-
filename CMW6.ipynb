2. Compute matrix-vector multiplication using loops and compare with np.dot().

```python
import numpy as np

def dot_product(vec1, vec2):

    vec1 = vec1.flatten()
    vec2 = vec2.flatten()
    return np.sum(np.multiply(vec1, vec2))

def matrix_vector_product(mat, vec):
    m, n = mat.shape
    p, q = vec.shape

    assert n == p, "The matrix and vector are not compatible for multiplication!"

    result = np.zeros((m, q), dtype=float)

    for i in range(m):
        result[i] = dot_product(mat[i, :], vec)

    return result

mat1 = np.full((3, 3), 3, dtype=int)
vec1 = np.full((3, 1), 2, dtype=int)

print("Matrix = \n", mat1)
print("Vector = \n", vec1)

result1 = matrix_vector_product(mat1, vec1)
print("Result1 (Manual) = \n", result1)

result2 = mat1 @ vec1
print("Result2 (NumPy @) = \n", result2)
```

3. Python function that performs Matrix-Matrix
product.

```python
def matrix_matrix_product(matA, matB):
    m, n = matA.shape
    p, q = matB.shape

    assert n == p, "Matrices are not compatible for multiplication!"

    result = np.zeros((m, q), dtype=float)

    for i in range(m):
        for j in range(q):
            result[i, j] = dot_product(matA[i, :], matB[:, j])

    return result

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

res_manual = matrix_matrix_product(A, B)

res_numpy = A @ B

print("Manual Result:\n", res_manual)
print("NumPy Result:\n", res_numpy)
```

 4. Python function that checks if it is a symmetric matrix using transpose of matrix.

```python
import numpy as np

def is_symmetric(matrix):
    arr = np.array(matrix)

    if arr.shape[0] != arr.shape[1]:
        return False

    return np.array_equal(arr, arr.T)

symmetric_matrix = [[1, 2, 3],
                    [9, 4, 5],
                    [3, 5, 6]]

print(is_symmetric(symmetric_matrix))
```
