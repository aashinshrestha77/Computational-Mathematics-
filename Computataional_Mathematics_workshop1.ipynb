```python
def fib():
    n = int(input("Enter number of terms: "))

    a = 0
    b = 1
    count = 0

    while count < n:
        print(a, end=" ")
        c = a + b
        a = b
        b = c
        count += 1

fib()
```

```python
def power(x, n):
    result = 1
    count = 0

    while count < n:
        result = result * x
        count += 1

    return result
print (power(5, 3))
```

```python
def sum(n):
    total = 0

    while n > 0:
        digit = n % 10
        total = total + digit
        n = n // 10

    print(total)
sum (79)
```

```python
def square(arr):
    total = 0
    i = 0

    while i < len(arr):
        total = total + arr[i] * arr[i]
        i += 1

    return total
print(square([5,3,4]))
```

```python
def sum(num):

    fact = 1
    i = 1
    while i <= num:
        fact = fact * i
        i += 1

    total = 0
    while fact > 0:
        total = total + (fact % 10)
        fact = fact // 10

    return total
print(sum(9))
```
